```python
def validationmetrics_reg(model,X_test,y_test, verbose=True):
    predictions = model.predict(X_test)
    # R-squared
    r2 = r2_score(y_test,predictions)
    # Adjusted R-squared
    r2_adjusted = 1-(1-r2)*(X_test.shape[0]-1)/(X_test.shape[0]-X_test.shape[1]-1)
    # MSE
    mse = mean_squared_error(y_test,predictions)
    #RMSE
    rmse = math.sqrt(mse)
    if verbose:
        print("R-Squared Value: ", r2)
        print("Adjusted R-Squared: ", r2_adjusted)
        print("RMSE: ", rmse)
    
    res_map = {
            "r2": r2,
            "r2_adjusted": r2_adjusted,
            "rmse": rmse,
            "model_obj": model
              }
    return res_map
```
